```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import decomposition
```

```python
from sklearn import datasets
```

```python
import mpl_toolkits.mplot3d
```

```python
np.random.seed(5)
```

```python
iris=datasets.load_iris()
```

```python
X=iris.data
y=iris.target
```

```python
fig = plt.figure(1, figsize=(8, 6))
plt.clf()
```

```python
ax = fig.add_subplot(111, projection='3d', elev="3d", azim=134)
```

```python
ax.set_position([0, 0, 0.95, 1])
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d', elev=48, azim=134)
ax.set_position([0, 0, 0.95, 1])

pca = decomposition.PCA(n_components=3)
pca.fit(X)
X_transformed = pca.transform(X)

# Store names and labels in a list for easier access
class_info = [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]

for name, label in class_info:
    ax.text3D(X_transformed[y == label, 0].mean(),
              X_transformed[y == label, 1].mean() + 1.5,
              X_transformed[y == label, 2].mean(),
              name,
              horizontalalignment="center",
              bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"))

# Plot the points for each class separately
colors = ['r', 'g', 'b'] # Define colors for each class
for (name, label), color in zip(class_info, colors):
    ax.scatter(X_transformed[y == label, 0],
               X_transformed[y == label, 1],
               X_transformed[y == label, 2],
               c=color,
               edgecolor="k",
               label=name) # Use the name directly for the legend

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

ax.set_title("PCA on Iris Dataset (3D)")
ax.set_xlabel("1st eigenvector")
ax.set_ylabel("2nd eigenvector")
ax.set_zlabel("3rd eigenvector")
ax.legend()

plt.show()
```
